```python
# 1. Install widget support if needed
!pip install ipywidgets --quiet

# 2. Main app code
import ipywidgets as widgets
from IPython.display import display, clear_output
import json

todo_list = []
def load_tasks():
    global todo_list
    try:
        with open('todo_list.txt', 'r') as f:
            todo_list = json.load(f)
    except Exception:
        todo_list = []

def save_tasks():
    with open('todo_list.txt', 'w') as f:
        json.dump(todo_list, f)

task_input = widgets.Text(
    placeholder='Enter a new task',
    description='Task:',
    disabled=False
)
add_button = widgets.Button(description='Add Task', button_style='success')
tasks_box = widgets.VBox([])
output = widgets.Output()

def refresh_tasks():
    tasks_widgets = []
    for idx, task in enumerate(todo_list):
        del_btn = widgets.Button(description='Delete', button_style='danger', layout=widgets.Layout(width='70px'))
        cb = widgets.Checkbox(value=task['done'], description=task['title'], indent=False)

        def on_checkbox_change(change, idx=idx):
            if change['name'] == 'value':
                todo_list[idx]['done'] = change['new']
                save_tasks()
        cb.observe(on_checkbox_change, names='value')

        def on_delete_click(b, idx=idx):
            del todo_list[idx]
            save_tasks()
            refresh_tasks()
        del_btn.on_click(lambda b, idx=idx: on_delete_click(b, idx))
        tasks_widgets.append(widgets.HBox([cb, del_btn]))
    tasks_box.children = tasks_widgets

def on_add_clicked(b):
    desc = task_input.value.strip()
    if desc:
        todo_list.append({'title': desc, 'done': False})
        task_input.value = ''
        save_tasks()
        refresh_tasks()
add_button.on_click(on_add_clicked)

def run_app():
    load_tasks()
    refresh_tasks()
    display(widgets.HBox([task_input, add_button]))
    display(tasks_box)
    display(output)
    with output:
        clear_output()
        print('Simple To-Do List App ready!')

run_app()
```
